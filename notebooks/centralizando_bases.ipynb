```python
import pandas as pd
import unidecode
```

## Funções auxiliares de tratamento de acentuação e uppercase

```python
# Função para remover a acentuação
def remove_accentuation(df):
    return df.apply(lambda item: unidecode.unidecode(item))

# Função para deixar os termos em letra maiúscula 
def to_uppercase(df):
    return df.apply(lambda item: item.strip().upper())

# Função para traduzir o formato de data para o mesmo da base principal
def translate_date(date):
    x = pd.to_datetime(date)
    try:
        new_date = "{:02d}/{:02d}/{:04d}".format(int(x.day), int(x.month), int(x.year))
    except:
        new_date = None
    return new_date
```

## Dados leitos e ocupação

```python
# DataSUS ocupacao leito e ocupacao
df_leito_ocupacao = pd.read_csv('Bases_cruzadas/esus-vepi.LeitoOcupacao.csv', sep=',')
df_leito_ocupacao.head(2)
df_leito_ocupacao['ID_MUNICIP'] = remove_accentuation(df_leito_ocupacao['municipioNotificacao'])
df_leito_ocupacao['ID_MUNICIP'] = to_uppercase(df_leito_ocupacao['ID_MUNICIP'])
df_leito_ocupacao['DT_NOTIFIC'] = df_leito_ocupacao['dataNotificacao'].apply(lambda x: translate_date(x))
df_leito_ocupacao
```


## Centralizando os dados

### Juntando as informações do df_leito_ocupação no df principal (main_df) pela DT_NOTIFIC e ID_MUNICIP

```python
df_leito_ocupacao_merge = df_leito_ocupacao[
    ['ID_MUNICIP',
     'DT_NOTIFIC',
     'ocupacaoSuspeitoCli',
     'ocupacaoSuspeitoUti',
     'ocupacaoConfirmadoCli',
     'ocupacaoConfirmadoUti',
     'saidaSuspeitaObitos',
     'saidaSuspeitaAltas',
     'saidaConfirmadaObitos',
     'saidaConfirmadaAltas']]
df_main = df_main.merge(df_leito_ocupacao, on = ['ID_MUNICIP', 'DT_NOTIFIC'], how='left')
```

### Gerando versao apenas com informações relevantes dos demais datasets

```python
df_agencia_ibge_enfermeiros_merge = df_agencia_ibge_enfermeiros[
    ['ID_MUNICIP',
     'População Estimada (2019)',
     'Total de Enfermeiros',
     'Enfermeiros do SUS',
     'Enfermeiros a cada 100 mil habitantes',
     'Enfermeiros do SUS a cada 100 mil habitantes']]

df_agencia_ibge_medicos_merge = df_agencia_ibge_medicos[
    ['ID_MUNICIP',
     'Total de médicos',
     'Médicos do SUS',
     'Médicos a cada 100 mil habitantes',
     'Médicos do SUS a cada 100 mil habitantes']]

df_agencia_ibge_respiradores_merge = df_agencia_ibge_respiradores[
    ['ID_MUNICIP',
     'Total de Respiradores',
     'Respiradores a cada 100 mil habitantes',
     'Respiradores do SUS a cada 100 mil habitantes',
     'Total de Respiradores do SUS']]

df_agencia_ibge_leitos_merge = df_agencia_ibge_leitos[
    ['ID_MUNICIP',
     'Total de Leitos de UTI',
     'Leitos de UTI a cada 100 mil habitantes',
     'Leitos de UTI do SUS',
     'Leitos do SUS a cada 100 mil habitantes']]
```

### Fazendo merge dos datasets entre si antes do merge com o princial (df_main)

```python
df_merged_ibge = df_agencia_ibge_enfermeiros_merge.merge(df_agencia_ibge_medicos_merge, on ='ID_MUNICIP', how='left')
df_merged_ibge = df_merged_ibge.merge(df_agencia_ibge_respiradores_merge, on ='ID_MUNICIP', how='left')
df_merged_ibge = df_merged_ibge.merge(df_agencia_ibge_leitos_merge, on ='ID_MUNICIP', how='left')
df_merged_ibge.head(2)
```

### Merge com dataset principal

```python
df_principal = df_main.merge(df_merged_ibge, on='ID_MUNICIP', how='left')
```

### Salvando em arquivo csv

```python
df_merged_ibge.to_csv("df_merged_ibge_only.csv", index=False)
df_principal.to_csv("df_principal_centralizado.csv", index=False)
df_main.to_csv("df_principal_ocupacao_leito.csv", index=False)
```
