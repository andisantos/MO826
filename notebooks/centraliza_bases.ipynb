```python
# Função para remover a acentuação
def remove_accentuation(df):
    return df.apply(lambda item: unidecode.unidecode(item))

# Função para deixar os termos em letra maiúscula 
def to_uppercase(df):
    return df.apply(lambda item: item.strip().upper())

# Função para traduzir o formato de data para o mesmo da base principal
def translate_date(date):
    x = pd.to_datetime(date)
    try:
        new_date = "{:02d}/{:02d}/{:04d}".format(int(x.day), int(x.month), int(x.year))
    except:
        new_date = None
    return new_date
  
uf_dict = {
    'ACRE': 'AC',
    'ALAGOAS': 'AL',
    'AMAPA': 'AP',
    'AMAZONAS': 'AM',
    'BAHIA': 'BA',
    'CEARA': 'CE',
    'DISTRITO FEDERAL': 'DF',
    'ESPIRITO SANTO': 'ES',
    'GOIAS': 'GO',
    'MARANHAO': 'MA',
    'MATO GROSSO': 'MT',
    'MATO GROSSO DO SUL': 'MS',
    'MINAS GERAIS': 'MG',
    'PARA': 'PA',
    'PARAIBA': 'PB',
    'PARANA':'PR',
    'PERNAMBUCO': 'PE',
    'PIAUI': 'PI',
    'RIO DE JANEIRO': 'RJ',
    'RIO GRANDE DO NORTE': 'RN',
    'RIO GRANDE DO SUL': 'RS',
    'RONDONIA': 'RO',
    'RORAIMA': 'RR',
    'SANTA CATARINA': 'SC',
    'SAO PAULO': 'SP',
    'SERGIPE': 'SE',
    'TOCANTINS': 'TO'
}

def translate_uf(df):
  return df.apply(lambda x: uf_dict[x])
```

```python
# Bases Agencia IBGE
df_agencia_ibge_leitos = pd.read_excel(medicos_enfermeiros_path, sheet_name = 'Leitos_de_UTI')
df_agencia_ibge_respiradores = pd.read_excel(medicos_enfermeiros_path, sheet_name = 'Respiradores_por_municípios')
df_agencia_ibge_medicos = pd.read_excel(medicos_enfermeiros_path, sheet_name = 'Medicos_por_municípios')
df_agencia_ibge_enfermeiros = pd.read_excel(medicos_enfermeiros_path, sheet_name = 'Enfermeiros_municípios')
```

## Dados de respiradores

```python
df_agencia_ibge_respiradores['ID_MN_RESI'] = remove_accentuation(df_agencia_ibge_respiradores['Município'])
df_agencia_ibge_respiradores['ID_MN_RESI'] = to_uppercase(df_agencia_ibge_respiradores['ID_MN_RESI'])
df_agencia_ibge_respiradores['SG_UF'] = remove_accentuation(df_agencia_ibge_respiradores['Estado ou Distrito Federal'])
df_agencia_ibge_respiradores['SG_UF'] = to_uppercase(df_agencia_ibge_respiradores['SG_UF'])
df_agencia_ibge_respiradores['SG_UF'] = translate_uf(df_agencia_ibge_respiradores['SG_UF'])
df_agencia_ibge_respiradores.head(2)
```

```python
df_agencia_ibge_medicos['ID_MN_RESI'] = remove_accentuation(df_agencia_ibge_medicos['Município'])
df_agencia_ibge_medicos['ID_MN_RESI'] = to_uppercase(df_agencia_ibge_medicos['ID_MN_RESI'])
df_agencia_ibge_medicos['SG_UF'] = remove_accentuation(df_agencia_ibge_medicos['Estado'])
df_agencia_ibge_medicos['SG_UF'] = to_uppercase(df_agencia_ibge_medicos['SG_UF'])
df_agencia_ibge_medicos['SG_UF'] = translate_uf(df_agencia_ibge_medicos['SG_UF'])
df_agencia_ibge_medicos.head(2)
```

## Dados de enfermeiros

```python
df_agencia_ibge_enfermeiros['ID_MN_RESI'] = remove_accentuation(df_agencia_ibge_enfermeiros['Município'])
df_agencia_ibge_enfermeiros['ID_MN_RESI'] = to_uppercase(df_agencia_ibge_enfermeiros['ID_MN_RESI'])
df_agencia_ibge_enfermeiros['SG_UF'] = remove_accentuation(df_agencia_ibge_enfermeiros['Estado ou Distrito Federal'])
df_agencia_ibge_enfermeiros['SG_UF'] = to_uppercase(df_agencia_ibge_enfermeiros['SG_UF'])
df_agencia_ibge_enfermeiros['SG_UF'] = translate_uf(df_agencia_ibge_enfermeiros['SG_UF'])
df_agencia_ibge_enfermeiros.head(2)
```

## Dados leitos e ocupação

```python
# DataSUS ocupacao leito e ocupacao
df_leito_ocupacao = pd.read_csv(ocupacao_leito_path, sep=',')
df_leito_ocupacao['ID_MN_RESI'] = remove_accentuation(df_leito_ocupacao['municipioNotificacao'])
df_leito_ocupacao['ID_MN_RESI'] = to_uppercase(df_leito_ocupacao['ID_MN_RESI'])
df_leito_ocupacao['SG_UF'] = remove_accentuation(df_leito_ocupacao['estadoNotificacao'])
df_leito_ocupacao['SG_UF'] = to_uppercase(df_leito_ocupacao['SG_UF'])
df_leito_ocupacao['SG_UF'] = translate_uf(df_leito_ocupacao['SG_UF'])
df_leito_ocupacao['DT_NOTIFIC'] = df_leito_ocupacao['dataNotificacao'].apply(lambda x: translate_date(x))
df_leito_ocupacao.head(2)
```

## Dados Trabalhadores Informais

```python
df_ibge_trabalhadores_informais = pd.read_excel(trabalhadores_informais_path, sheet_name = 'trabalhadores_informais_for_mer')
df_ibge_trabalhadores_informais['SG_UF'] = remove_accentuation(df_ibge_trabalhadores_informais['UF'])
df_ibge_trabalhadores_informais['SG_UF'] = to_uppercase(df_ibge_trabalhadores_informais['SG_UF'])
df_ibge_trabalhadores_informais['SG_UF'] = translate_uf(df_ibge_trabalhadores_informais['SG_UF'])
df_ibge_trabalhadores_informais.head(27)
```


## Centralizando os dados

### Gerando versao apenas com informações relevantes de todos os datasets

```python
df_leito_ocupacao_merge = df_leito_ocupacao[
    ['ID_MN_RESI',
     'SG_UF',
     'DT_NOTIFIC',
     'ocupacaoSuspeitoCli',
     'ocupacaoSuspeitoUti',
     'ocupacaoConfirmadoCli',
     'ocupacaoConfirmadoUti',
     'saidaSuspeitaObitos',
     'saidaSuspeitaAltas',
     'saidaConfirmadaObitos',
     'saidaConfirmadaAltas']]

df_agencia_ibge_enfermeiros_merge = df_agencia_ibge_enfermeiros[
    ['ID_MN_RESI',
     'SG_UF',
     'População Estimada (2019)',
     'Total de Enfermeiros',
     'Enfermeiros do SUS',
     'Enfermeiros a cada 100 mil habitantes',
     'Enfermeiros do SUS a cada 100 mil habitantes']]

df_agencia_ibge_medicos_merge = df_agencia_ibge_medicos[
    ['ID_MN_RESI',
     'SG_UF',
     'Total de médicos',
     'Médicos do SUS',
     'Médicos a cada 100 mil habitantes',
     'Médicos do SUS a cada 100 mil habitantes']]

df_agencia_ibge_respiradores_merge = df_agencia_ibge_respiradores[
    ['ID_MN_RESI',
     'SG_UF',
     'Total de Respiradores',
     'Respiradores a cada 100 mil habitantes',
     'Respiradores do SUS a cada 100 mil habitantes',
     'Total de Respiradores do SUS']]

df_agencia_ibge_leitos_merge = df_agencia_ibge_leitos[
    ['ID_MN_RESI',
     'SG_UF',
     'Total de Leitos de UTI',
     'Leitos de UTI a cada 100 mil habitantes',
     'Leitos de UTI do SUS',
     'Leitos do SUS a cada 100 mil habitantes']]
    
df_ibge_trabalhadores_informais_merge = df_ibge_trabalhadores_informais[
    ['SG_UF',
     'Trabalhadores informais na unidade da Federação (julho 2019)',
     'Trabalhadores informais nas Regiões Metropolitas (julho 2019)',
     'Trabalhadores informais Capital (julho 2019)',
     'Trabalhadores informais no Interior (julho 2019)'
    ]
]
```

### Fazendo merge dos datasets entre si antes do merge com o principal

```python
df_merged_ibge = df_agencia_ibge_enfermeiros_merge.merge(df_agencia_ibge_medicos_merge, on =['ID_MN_RESI', 'SG_UF'], how='left')
df_merged_ibge = df_merged_ibge.merge(df_agencia_ibge_respiradores_merge, on =['ID_MN_RESI', 'SG_UF'], how='left')
df_merged_ibge = df_merged_ibge.merge(df_agencia_ibge_leitos_merge, on =['ID_MN_RESI', 'SG_UF'], how='left')
df_merged_ibge = df_merged_ibge.merge(df_ibge_trabalhadores_informais_merge, on =['SG_UF'], how='left')
df_merged = df_leito_ocupacao_merge.merge(df_merged_ibge, on = ["ID_MN_RESI", "SG_UF"], how = 'left')
df_merged.head(27)
```

### Salvando em arquivo csv

```python
df_merged.to_csv("/content/drive/MyDrive/Colab Notebooks/merged_datasets_datasus_ibgenoticias_v5.csv", index=False)
```
